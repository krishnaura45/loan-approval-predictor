# Model Performances
calculate the oof scores of each model and sort the models 

```python
model_names = [col for col in oofs.columns if col != 'id']
auc_scores = {}
for model in model_names:
    auc = roc_auc_score(train['loan_status'], oofs[model])
    auc_scores[model] = auc
    print(f"{model} AUC: {auc:.4f}")
```

```python
sorted_models = sorted(auc_scores.items(), key=lambda x: x[1], reverse=True)
print("Sorted Models by AUC:")
for model, auc in sorted_models:
    print(f"{model}: {auc:.4f}")
```

# Hill climbing approach

```python
current_best_ensemble = oofs[sorted_models[0][0]]
current_best_auc = roc_auc_score(train['loan_status'], current_best_ensemble)

weights = {sorted_models[0][0]: 1.0}
weight_range = np.arange(-0.1, 0.51, 0.01)


for _, (model_name, _) in enumerate(sorted_models[1:]): 
    potential_new_best_auc = current_best_auc
    best_model, best_weight = None, None
    
    for weight in weight_range:
        potential_ensemble = (1 - weight) * current_best_ensemble + weight * oofs[model_name]
        auc = roc_auc_score(train['loan_status'], potential_ensemble)
        
        if auc > potential_new_best_auc:
            potential_new_best_auc = auc
            best_model, best_weight = model_name, weight

    if best_model:
        print(f"Best Model: {best_model}, Weight: {best_weight}, AUC: {potential_new_best_auc:.4f}")
        current_best_ensemble = (1 - best_weight) * current_best_ensemble + best_weight * oofs[best_model]
        current_best_auc = potential_new_best_auc
        
        for model, weight in weights.items():
            weights[model] = weights[model] * (1 - best_weight)
        
        weights[best_model] = best_weight

print(f"Final Best AUC: {current_best_auc:.4f}")
```

```python
print("Model Weights in Final Ensemble:")
for model, weight in weights.items():
    print(f"{model}: {weight}")
```

# Submission Time

```python
final_submission = pd.Series(0, index=preds.index)  
for model, weight in weights.items():
    final_submission += weight * preds[model] 


submission = pd.DataFrame({'id': preds['id'], 'loan_status': final_submission})
submission.to_csv('submission.csv', index=False)

submission.head()
```
